```python
# Extract the months from the dataset and store in a separate column. 

# I will split the dates.

months = customer_case_df["Effective To Date"].str.split("/")
```

```python
months
```

```python
# Here I will extract the months from the dates and save them in a separate list.

month = []

def months_to_list(x):
    for i in x:
        month.append(i[0])
```

```python
months_to_list(months)
```

```python
month
```

```python
# I will make a new column with the months I extracted.

customer_case_df["Effective To Date Month"] = month
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# I will check the normality column by column.

sns.displot(data = customer_case_df, x = "Customer Lifetime Value", kde = True)
sns.displot(data = customer_case_df, x = "Income", kde = True)
sns.displot(data = customer_case_df, x = "Monthly Premium Auto", kde = True)
sns.displot(data = customer_case_df, x = "Months Since Last Claim", kde = True)
sns.displot(data = customer_case_df, x = "Months Since Policy Inception", kde = True)
sns.displot(data = customer_case_df, x = "Number of Open Complaints", kde = True)
sns.displot(data = customer_case_df, x = "Number of Policies", kde = True)
sns.displot(data = customer_case_df, x = "Total Claim Amount", kde = True)
```

```python
# Same as previous, just looks better. Thank you, Roman!

def distrib_plots(num_df):
    plt.figure(figsize = (12,10))
    for i, col in enumerate(num_df.columns, 1):
        plt.subplot(3,3,i)
        sns.histplot(num_df[col])
```

```python
# Plot with the total number of responses.

customer_case_df["Response"].value_counts().plot(kind='bar')
plt.title("Responses", fontsize=14)
plt.xlabel("Response", labelpad=10)
plt.ylabel("Nr. of responses", labelpad=10)
plt.ylim([0, 9000])
```

```python
# Plot with the response rate by the sales channel.

sns.countplot(x = customer_case_df["Response"], hue = customer_case_df["Sales Channel"], palette = "Set3")
```

```python
# Plot with the response rate by the total claim amount.
# Boxplot for the continuos variable.

sns.boxplot(x = customer_case_df["Response"], y = customer_case_df["Total Claim Amount"])
            
```

```python
# Plot with the response rate by income.

sns.boxplot(x = customer_case_df["Response"], y = customer_case_df["Income"])
```

```python
# Here is the correlation heatmap.

correl = numerical.corr()

mask = np.zeros_like(correl)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(9, 7))
    ax = sns.heatmap(correl, mask=mask,cmap='coolwarm', vmin=-1,vmax=1,annot=True, square=True)
```

```python
# I will explore my data further, by making a few more plots.

# The first one is a jointplot.

first_plot = sns.jointplot(data=customer_case_df,x="Total Claim Amount", y="Customer Lifetime Value", kind = "scatter")
first_plot.fig.suptitle("Relationship between Total claim amount and Customer lifetime value")
```

```python
# Here is a stripplot, that shows the relationship between Total claim amount and Gender and Education.

fig_explore_1, ax4 = plt.subplots(figsize = (14,10))
ax4.set_title("Total claim amount by gender and education", size='25', fontweight='bold')

sns.stripplot(x = customer_case_df["Gender"], y = customer_case_df["Total Claim Amount"], hue = customer_case_df["Education"], size = 8)
```

```python
# The last violin plot shows relationship between Income and Number of Open Complaints.

sns.violinplot(y="Income", x="Number of Open Complaints", data=customer_case_df)
```
